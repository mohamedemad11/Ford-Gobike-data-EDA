# Univariate Explorations


### he next 4 cells are about making functions in order to reducing the number of code lines

```python
#a function to make countplot 
def countplot(x,title):
    plt.figure(figsize=(20,10))
    color=sns.color_palette()[0]
    sns.countplot(x,color=color) 
    plt.title(str(title),fontsize=18)
    plt.show()
```

```python
#a function to make boxplot
def boxplot(x,title):
    plt.figure(figsize=(20,10))
    color=sns.color_palette()[0]
    sns.boxplot(x,color=color) 
    plt.title(str(title),fontsize=18)
    plt.show()
```

```python
#a function to make distripution plot
def distplot(x,title):
    plt.figure(figsize=(20,10))
    color=sns.color_palette()[0]
    sns.distplot(x,color=color) 
    plt.title(str(title),fontsize=18)
    plt.show()
```

```python
#a function to make violinplot
def violinplot(x,title):
    plt.figure(figsize=(20,10))
    color=sns.color_palette()[0]
    sns.violinplot(x,color=color)
    plt.title(str(title),fontsize=18)
    plt.show()
```

```python
#Building a function to automate bivariate analysis :
def countplot_bi(x,hue,title):
    plt.figure(figsize=(20,10)) 
    sns.countplot(x,hue=hue)
    plt.title(str(title),fontsize=18)
    plt.xticks(rotation=0)
    plt.show()
```

# What is the most frequent start hour of trips by gender_segmentation ?


```python
countplot_bi(data.start_hour,data.member_gender,'frequency of each trip start hour by gender')
```

# Observation :
### Most of our males and females user start their trip from 7 to 9 am  & 4 to 6 pm 

# what are the most frequent start hours for our subscribers and our users ?

```python
countplot_bi(data.start_hour,data.user_type,'frequency of each trip start hour by user type ')
```

### Observation :
### The peak time for subscribers is 8 am and 5 pm for customers 

# What is the most frequent weekday trip per member gender ?

# Multivariable explorations

```python
# A function to automate multivariable explorations :
def barplot_multi(x,y,hue,title):
    plt.figure(figsize=(20,10))
    sns.barplot(x,y,hue=hue)
    plt.title(str(title),fontsize=18)
    plt.xticks(rotation=0)
    plt.show()
```

# Distribution of User Age & Durations minutes per member gender :

```python
barplot_multi(data.user_age,data.duration_min,data.member_gender,'Duration vs age by user gender')
```

# Observation :
### most of our customers are 20 years old and about 15 minutes per trip 


# which type of users tend to last longer in the trips ?

```python
barplot_multi(data.user_age,data.duration_min,data.user_type,'Duration vs Age per user type')
```

# Observation :
### customers tend to last longer in trips than subscribers 

# Durations per minute of weekdays per user type :

```python
barplot_multi(data.day_of_week_start,data.duration_min,data.user_type,'Duration vs Weekday per user type')
```

# Which Member gender tend to last longer in trips :

```python
barplot_multi(data.day_of_week_start,data.duration_min,data.member_gender,'Duration vs Weekday per user Gender')
```

# Observation
### females tend to last longer in all of the weekdays
