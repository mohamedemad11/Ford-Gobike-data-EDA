```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
import pandas as pd
%matplotlib inline 
```

```python
data=pd.read_csv('bikes_cleaned.csv',)
```

```python
data.head()
```

```python
data=data.iloc[:,1:]
```

```python
data.head()
```

```python
data.info()
```

```python
data['start_time']=pd.to_datetime(data['start_time'])
data['end_time']=pd.to_datetime(data['end_time'])
```

```python
data.info()
```

```python
plt.figure(figsize=(14,7))
sns.countplot(data=data,x='start_hour',color=sns.color_palette()[0])
plt.xlabel('Start_Hour',fontsize=20)
plt.ylabel('Count',fontsize=20)
plt.title('Hours trips Peaks',fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.countplot(data=data,x='day_of_week_start',color=sns.color_palette()[0])
plt.xticks(rotation=15)
plt.title('Most days of Weeks Trips',fontsize=20)
plt.xlabel('days of WEEK',fontsize=20)
plt.ylabel('Count',fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(14,7))
data.groupby('member_gender')['duration_min'].sum().plot(kind='bar')
plt.ylabel('total duration by mins ',fontsize=14)
plt.title('Which member gender lasts longer',fontsize=20)
plt.xlabel('member Gender',fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(14,7))
sns.countplot(data.member_gender)
plt.title('user gender count',fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(14,7))
data.groupby('member_gender')['duration_min'].mean().plot(kind='bar')
plt.ylabel('Avg duration by mins ',fontsize=14)
plt.title('Which member gender lasts longer By Average',fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(20,10))
data.member_gender.value_counts(normalize=True).plot(kind='pie',autopct='%1.1f%%',shadow=True,fontsize=14)
plt.title('our members gender percentage',fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(20,10))
data.groupby(['member_gender','user_type'])['duration_min'].count().sort_values().plot(kind='barh')
plt.xticks(rotation=15,fontsize=16)
plt.yticks(fontsize=16)
plt.title('The relationship between member_gender , user_type , Number of rides ',fontsize=20)
plt.xlabel('Number of Trips',fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(20,10))
data.groupby(['member_gender','user_type'])['duration_min'].mean().sort_values().plot(kind='barh')
plt.xticks(rotation=15,fontsize=16)
plt.yticks(fontsize=16)
plt.title('The relationship between member_gender , user_type , avg_duration',fontsize=20)
plt.xlabel('Duration Mins',fontsize=16)
plt.show()
```
